```python
import scanpy as sc

# 读取 h5ad 文件
adata = sc.read_h5ad("/data/zhaoyh/DeSCENT/data/BRCA/single_cell/BRCA_train.symbol_mapped.h5ad")

# 查看基本信息
print(adata)
print("细胞数, 基因数:", adata.shape)

# 查看细胞注释信息 obs
print("\nobs（细胞信息）:")
print(adata.obs.head())
print("obs 列名:", adata.obs.columns.tolist())

# 查看基因信息 var
print("\nvar（基因信息）:")
print(adata.var.head())
print("var 列名:", adata.var.columns.tolist())

# 查看表达矩阵 X
print("\n表达矩阵 X:")
print(type(adata.X))
print("矩阵形状:", adata.X.shape)

# 查看前5个细胞、前5个基因的表达值
print("\n前5×5表达矩阵:")
print(adata.X[:5, :5].toarray() if hasattr(adata.X, "toarray") else adata.X[:5, :5])
```
